```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("stocks.sqlite")
conn
```

```python
cur = conn.cursor()
cur
```

```python
res = cur.execute("SELECT name FROM sqlite_master WHERE type='table'")
res.fetchall()
```

```python
res = cur.execute("SELECT * FROM STOCK_DATA WHERE Symbol = 'MSFT'")
msft_data = res.fetchall()
msft_data
```

```python
res = cur.execute("SELECT * FROM STOCK_DATA WHERE Symbol = 'AAPL'")
aapl_data = res.fetchall()
aapl_data
```

```python
res = cur.execute("PRAGMA table_info(STOCK_DATA)")
col_names = [x[1] for x in res.fetchall()]
col_names
```

```python
",".join(col_names)
```

```python
['?']* len(col_names)
```

```python
res =cur.execute("DROP TABLE msft")
```

```python
# Escape column names to handle reserved keywords
escaped_col_names = [f'"{name}"' for name in col_names]

# Create new tables with the same schema as STOCK_DATA
cur.execute(f"CREATE TABLE msft({','.join(escaped_col_names)})")
cur.executemany(f"INSERT INTO msft VALUES({','.join(['?']*len(col_names))})", msft_data)
```

```python
res =cur.execute("DROP TABLE AAPL")
```

```python
# Escape column names to handle reserved keywords
escaped_col_names = [f'"{name}"' for name in col_names]

# Create new tables with the same schema as STOCK_DATA
cur.execute(f"CREATE TABLE AAPL({','.join(escaped_col_names)})")
cur.executemany(f"INSERT INTO AAPL VALUES({','.join(['?']*len(col_names))})", aapl_data)
```
