```python
res = cur.execute("SELECT name FROM sqlite_master WHERE type='table'")
res.fetchall()
```

```python
pd.read_sql("""SELECT  EmployeeNumber, Department, Age, Gender, Attrition
            FROM employee
            WHERE Department = 'Sales'""",con)
```

```python
res = cur.execute("""SELECT  EmployeeNumber, Department, Age, Gender, Attrition
                  FROM employee
                  WHERE Department = 'Sales'""")
data=res.fetchall()

res = cur.execute("PRAGMA table_info(employee)")

col_names = ["EmployeeNumber", "Department", "Age", "Gender", "Attrition"]
sales = pd.DataFrame(data, columns=col_names)
sales # 2
```

```python
res = cur.execute("""SELECT  EmployeeNumber, EducationField, Age, Gender, Attrition
                  FROM employee
                  WHERE EducationField = 'Life Sciences'""")
data=res.fetchall()

col_names =["EmployeeNumber", "EducationField", "Age", "Gender", "Attrition"]

field = pd.DataFrame(data, columns=col_names)
field # 3
```

```python
cur.execute("CREATE TABLE sales(EmployeeNumber,Department, Age, Gender, Attrition)")
cur.executemany("INSERT INTO sales VALUES(?,?,?,?,?)", sales.itertuples(index=False, name=None))
# cur.commit()
```
