```python
import sqlite3
import pandas as pd
```

```python
con = sqlite3.connect("/content/telecom.db")
con
```

```python
cur = con.cursor()
cur
```

```python
# 1
res = cur.execute("SELECT name FROM sqlite_master WHERE type='table'")
res.fetchall()
```

```python
res = cur.execute("PRAGMA table_info(train)")
col_names = [x[1] for x in res.fetchall()]
col_names
```

```python
res = cur.execute("PRAGMA table_info(event_type)")
col_names = [x[1] for x in res.fetchall()]
col_names
```

```python
res = cur.execute("PRAGMA table_info(severity_type)")
col_names = [x[1] for x in res.fetchall()]
col_names
```

```python
res = cur.execute("PRAGMA table_info(resource_type)")
col_names = [x[1] for x in res.fetchall()]
col_names
```

```python
res = cur.execute("PRAGMA table_info(log_feature)")
col_names = [x[1] for x in res.fetchall()]
col_names
```

```python
# 2
res1 = cur.execute("""CREATE TEMP TABLE joined_table AS
                    SELECT t.id, t.location, t.fault_severity, et.event_type, st.severity_type, rt.resource_type, lf.log_feature, lf.volume
	                FROM train t
	                LEFT JOIN event_type et  ON t.id = et.id
                    LEFT JOIN severity_type st ON et.id = st.id
                    LEFT JOIN resource_type rt ON st.id = rt.id
                    LEFT JOIN log_feature lf ON rt.id =lf.id;
                                                             """)
res1
```

```python
res1 =  pd.read_sql("""SELECT * FROM joined_table""",con)
res1
```

```python
# 3
res2 = pd.read_sql("""SELECT DISTINCT event_type, severity_type
                    FROM joined_table;""",con)
res2
```

```python
# 4
res3 = pd.read_sql("""SELECT "fault_severity", COUNT(*) as occurrence_count
                      FROM joined_table
                      GROUP BY "fault_severity" """, con)
res3
```

```python
cur.execute("""DROP TABLE joined_table""")
```
