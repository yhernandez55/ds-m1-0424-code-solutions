```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
file_path = '/content/Sales_data.txt'
df = pd.read_csv(file_path)
df
```

```python
df.describe()
```

# 1

```python
female = df[df['Contact Sex'] == 'F']
female_count = female.groupby(female['Contact Sex']).count()
female_count
```

```python
male = df[df['Contact Sex'] == 'M']
male_count = male.groupby(male['Contact Sex']).count()
male_count
```

Based on the female_count and male_count we can say that males are the most frequent buyers. You can see that the amount of females that made a perchase is 17 while the amount of buyers for males are 22.

# 2

Based on question 1, we can say that males should be targeted if the business is cash-constrained, since males were the more frequent buyers.

# 3

We can say that males should be targeted if consumers are cash-constrained, since they are more likely to purchase technololgy.



# 4

```python
age_group = df.groupby('Contact Age')['Profit'].sum().sort_values(ascending=False)
print(age_group)
```

After grouping by age, we can say that our target age to maximize profit should be 46, since the amount of purchases was higher for those who were 46 years old.

# 5

```python
product = df.groupby('Product Type')['Profit'].sum().sort_values(ascending=False)
product
```

After grouping the product types, we can see that laptops should be featured more, since more laptops are being sold compared to desktops and tablets.

# 6

```python
lead_source = df.groupby('Lead Source')['Profit'].sum().sort_values(ascending=False)
lead_source
```

The lead source that has worked in the past has been the website, since we can see that there was more profit coming in from that lead source compared to the other types.

# 7

```python
email_marketing = df[df['Lead Source'] == 'Email'].groupby('Sale Year')['Profit'].sum().sort_values(ascending=False)
email_marketing
```

The best time to do email marketing was in the year 2018, since more profit was made during that time compared to other years.
